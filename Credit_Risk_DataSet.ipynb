```python
#import pandas library 

import pandas as pd
```

```python
# load data into a pandas data frame
df = pd.read_csv("C:/Users/shash/Desktop/Portfolio Projects/credit_risk_dataset.csv")
```

```python
#The describe function displays the statistical summary of our data
df.describe()
```

```python
#Use info function to display the number of columns and its datatypes
df.info()
```

```python
#Use the dropna() function to drop the data with null values
df.dropna()
```

```python
#Use the sk learn library to split our data into training and testing sets.
#This is done as our data is imbalanced
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test = train_test_split(df,random_state=40)
```

```python
print(X_train.isnull().sum())
print(X_test.isnull().sum())
```

```python
#This process of using the fillna() function is known as imputing
#Here we replace null values with the mean of the data

X_train.fillna(X_train.mean(), inplace=True)
X_test.fillna(X_train.mean(), inplace=True)
```

```python
print(X_train.isnull().sum())
print(X_test.isnull().sum())
```

```python
#Categorical Data is converted into numeric
#Iterate through the data 
#Replace the Object data type values with the most frequently occuring numeric data in the column

for col in X_train.columns:
    if X_train[col].dtype == 'Object':
        X_train = X_train.fillna(X_train[col].value_counts().index[0])
        X_test = X_test.fillna(X_train[col].value_counts().index[0])
    
```

```python
print(X_train.isnull().sum())
print(X_test.isnull().sum())
```

```python
X_train.head()
X_test.head()
```

```python
The missing values are now successfully handled.

There is still some minor but essential data preprocessing needed before we proceed towards building our machine learning model. We are going to divide these remaining preprocessing steps into two main tasks:

    Convert the non-numeric data into numeric.
    Scale the feature values to a uniform range.

First, we will be converting all the non-numeric values into numeric ones. 
We do this because not only it results in a faster computation but also many machine learning models (like XGBoost) 
(and especially the ones developed using scikit-learn) require the data to be in a strictly numeric format.
We will do this by using the get_dummies() method from pandas.




X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

```python
X_train.head()
X_test.head()
```

```python
#Scaling our Data

from sklearn.preprocessing import MinMaxScaler

X_train, y_train = X_train.iloc[:, :-1].values, X_train.iloc[:, [-1]].values
X_test, y_test = X_test.iloc[:, :-1].values, X_test.iloc[:, [-1]].values


scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(rescaledX_train,y_train)
```

```python
# Import confusion_matrix
from sklearn.metrics import confusion_matrix

# Use logreg to predict instances from the test set and store it
y_pred = logreg.predict(rescaledX_test)

# Get the accuracy score of logreg model and print it
print("Accuracy of logistic regression classifier: ", logreg.score(rescaledX_test,y_test))

# Print the confusion matrix of the logreg model
confusion_matrix(y_test,y_pred)
```

```python
#Eventhough the Accuracy of logistic regression classifier was 1.0,
#Our dataset is imbalanceg and hence a better measure of accuracy is the Area under an ROC curve



from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
#define metrics
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
#define metrics
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```
